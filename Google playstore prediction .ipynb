```python
df_encoded=data2.copy()
```

```python
df_encoded
```

```python
df_encoded["Ad Supported id"]= le.fit_transform(df_encoded["Ad Supported"])
df_encoded
```

```python
category_counts=df_encoded["Category"].value_counts().to_dict()
df_encoded["categorie_count"]= df_encoded["Category"].map(category_counts)
df_encoded
```

```python
X= df_encoded[["categorie_count", "installation"]].values.astype(float)
Y= df_encoded["Applications"].values.astype(int)
```

```python
df_encoded["installation"]=le.fit_transform(df_encoded["Maximum Installs"])
df_encoded
```

```python
App_Names = df_encoded["App Name"].value_counts().to_dict()
df_encoded["Applications"]=df_encoded["App Name"].map(App_Names)
```
